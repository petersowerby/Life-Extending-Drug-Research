```python
# principle component analysis, reducing 1024 features to 2 (PC1 and PC2)

pca = PCA(n_components = 2)
print(X2.shape)
res = pca.fit_transform(X2)
print(res.shape)
```

```python
# creating a dataframe with PC1 and PC2 as the features

principle = pd.DataFrame(data = res, columns = ['PC1', 'PC2'])
principle.head()
```

```python
# adding 'Target' columns to dataframe

FinalPCA = pd.concat([principle, morgan2048[['Target']]], axis = 1)
FinalPCA.head()
```

```python
# visualising PC1 vs PC2

sns.set_style('dark')

colours = 'red', 'blue'
ax = sns.pairplot(data = FinalPCA, hue = 'Target', x_vars = ['PC1'], y_vars = ['PC2'], palette = colours, height = 5)

plt.xlabel('PC1', fontsize = 14)
plt.ylabel('PC2', fontsize = 14)
plt.title('Chemical Space (ECFPs 2048-bits)', fontsize = 16)
ax._legend.remove()
plt.legend(['Inactive = 1126','Active = 307'], fontsize = 12)
```
