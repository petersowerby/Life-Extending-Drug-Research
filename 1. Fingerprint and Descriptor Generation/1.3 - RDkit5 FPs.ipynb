# RDkit5 Fingerprint Generation

```python
# importing data and transforming to dataframe
df = pd.read_csv("final_df_smiles.csv", encoding = "ISO-8859-1")
df.head()
```

```python
# generating list of RDkit5 fingerprints
rdkbi = []

for i in range (0,len(df)):
    mol = Chem.MolFromSmiles(df.iloc[i,0])
    fp = Chem.RDKFingerprint(mol, maxPath=5)
    fp_list = np.unique(fp, return_inverse = True)[1].tolist()
    rdkbi.append(fp_list)
```

```python
# converting list to dataframe and adding output (active or inactive)
rdk5 = pd.DataFrame(rdkbi)
rdk5 = pd.concat([rdk5, df.iloc[:, -1]], axis = 1)
rdk5.head()
```

```python
# showing the number of compounds where target = 0 and target = 1
print("Inactive Compounds:{}".format(rdk5[(rdk5["Target"]==0)].shape[0]))
print("Active Compounds:{}".format(rdk5[(rdk5["Target"]==1)].shape[0]))
```
