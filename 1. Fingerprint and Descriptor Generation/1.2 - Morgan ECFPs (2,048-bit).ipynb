# Morgan Fingerprint Generation (2,048-bit)

```python
# importing data and tranforming to a dataframe
df = pd.read_csv("final_df_smiles.csv", encoding = "ISO-8859-1")
df.head()
```

```python
# generation of morgan fingerprints with 2048 bit length
morgan2048 = []

for i in range (0,len(df)):
    mol = Chem.MolFromSmiles(df.iloc[i,0])
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
    fp_list = np.unique(fp, return_inverse = True)[1].tolist()
    morgan2048.append(fp_list)
```

```python
# adding the output value to the dataframe (active or inactive)
morgan2048 = pd.DataFrame(data= morgan2048)
morgan2048 = pd.concat([morgan2048, df.iloc[:, -1]], axis = 1)
display(morgan2048.head())# showing the number of compounds where target = 0 and target = 1
print("Inactive Compounds:{}".format(morgan2048[(morgan2048["Target"]==0)].shape[0]))
print("Active Compounds:{}".format(morgan2048[(morgan2048["Target"]==1)].shape[0]))
```
